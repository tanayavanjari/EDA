**Creating 1D array:**
1D array is created using


1.   list of elements
2.   arange()
3.   linspace()
4.   zeros()
5.   ones()



```python
import numpy as np
```

```python
# Creating 1D array using python list
# let us crate a list of elements
x=[10,20,30,40,50] # List values are separated by comma
print(x)
print(type(x))

# Let us convert x into numpy array
a=np.array(x) # Array values are separated by spaces
print(a)
print(type(a))
```

```python
# Difference between List and array
# In list elements are separated by comma and in array elements are separated by space
# List is a python data structure, array is numpy data structure
# list is heterogeneous, array is homogeneous
# Array operations are very fast as compared to list
# Numpy array stores daa in continuous memory locations whereas list sotre data in random locations
# Numpy array are memory efficient
```

```python
# Element-wise addition of list elements
# check time of execution

x = [i for i in range(10000000)]
y = [i for i in range(10000000,20000000)]
z = []
import time
start = time.time()
for i in range(len(x)):
  z.append(x[i] + y[i])

print(time.time()-start)
```

```python
# addition of numpy array elements
# check time of execution. It is less as compared to list

x = np.arange(10000000)
y = np.arange(10000000,20000000)

start = time.time()
z = x + y
print(time.time()-start)
```

```python
# List
L = [i for i in range(10000000)]
import sys
sys.getsizeof(L)
```

```python
# Numpy array
# Memory used by Numpy array is less than list
N = np.arange(10000000,dtype = np.int16)
sys.getsizeof(N)
```

**arange():** It is used to create an array using sequence of numbers. This can be used in three ways.

**Attribute of Numpy Arrays:**



1.   **Dimensionality** : array_var_name.ndim
2.   **Size:** it gives total count of elements in an array - array_var_name.size
3.   **Shape:** it gives total rows and coumns in an array - arra_var_name.shape
4.   **DataType:** it gives data type of an array - array_var_name.dtype
5.   **itemsize:** it gives memory allocated per element in an array - arra_var_name.itemsize



```python
# Example - 2
# Let us create a 2D array
x=np.eye(4)
print("Identity Array")
print(x)

# Let us verify the dimensionality of x
print("No. of Dimensions: ",x.ndim)

# Let us verify the size of x
print("Total Elements: ",x.size)

# Let us verify the shape of x
print("Total Rows and Columns: ",x.shape)

# Let us verify the data type of an element in x
print("Datatype: ",x.dtype)

# Let us verify the itemsize of x
print("Memory: ",x.itemsize)
```

**Copying an Array:**
array_var_name.copy(): to copy the data of an array in another.

```python
# copying the data of an array into another array.
x1=np.array([[1,2,3],[4,5,6],[9,9,9]])
print(x1)
```

```python
# data in x1 is copied into y1
y1=x1.copy()  # Deep Copy
# A new array is created with same data at separate memory location
print(y1)
```

```python
print(x1 is y1)
```

```python
# change in y1 doesn't change x1
y1[0,0] = 500
print(x1)
```

```python
print(y1)
```

```python
# alternative method for copy
z1=x1   # shallow copy (aliasing). This is assignment
# No new array is created
# z1 and x1 point to the same memory
# This is just another name for the same array
print(z1)
```

```python
print(x1 is z1)
```

```python
# if you change element of z1, it is reflected to x1
z1[0,0] = 100
print(x1)
```
