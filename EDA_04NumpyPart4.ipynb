# **Numpy Data Types:**

NumPy, a popular Python library for numerical computing, supports various data types to facilitate array operations. Here are some of the commonly used data types in NumPy:

Integer Types

1.   **Integer Types:**




          (a) int8, int16, int32, int64: Signed integers of 8, 16, 32, and 64 bits.

          (b) uint8, uint16, uint32, uint64: Unsigned integers of 8, 16, 32, and 64 bits.

2.   **Floating-Point Types:**

          (a) float16, float32 (also called single)
          (b) float64 (also called double)

3.   **Complex Types:**

          (a) complex64 (also called single),
          (b) complex128 (also called double):
                Complex numbers with 32 and 64-bit floating-point numbers for real and imaginary parts.

4.   **Boolean Type:**

          bool_: Boolean values (True or False).

5.   **String Type:**

          str_: Fixed-length Unicode strings.

6.   **Object Type:**

          object_: General Python objects.

7.   **Datetime and Timedelta Types:**

          (a) datetime64: Dates and times.

          (a) timedelta64: Differences between dates and times

# **Creating a NumPy Array with a Specific Data Type:**

```python
import numpy as np
```

```python
# Creating Numpy Arrays with specific DataType:
a=np.array([1, 2, 3, 4], dtype=np.int32)
print(a)
print(a.dtype)
```

```python
# Float array
f=np.array([1, 2, 3, 4], dtype=np.float64)
print(f)
print(f.dtype)
```

```python
# Complex array
c=np.array([1, 2, 3], dtype=np.complex128)
print(c)
print(c.dtype)
```

```python
# Boolean array
b=np.array([1, 0, 2, 0], dtype=np.bool_)
print(b)
print(b.dtype)
```

```python
# Boolean array-2
b=np.array([2, 0, 23, 0], dtype=np.bool_)
print(b)
print(b.dtype)
```

```python
# String array
s=np.array([1, 2, 3], dtype=np.str_)
print(s)
print(s.dtype) #<U1 U   indicates unicode, 1  maximum length = 1 character
```

```python
s=np.array(['ab', 'b', 'c'], dtype=np.str_)
print(s)
print(s.dtype) # #<U1 U   indicates unicode, 2  maximum length = 2 character
```

```python
s=np.array([1, 22, 3], dtype=np.str_)
print(s)
print(s.dtype)
```

```python
# Zeros array with float data type
z=np.zeros((2, 3), dtype=np.float32)
print(z)
print(z.dtype)
```

```python
# Ones array with integer data type
n=np.ones((3, 3), dtype=np.int16)
print(n)
print(n.dtype)
```

```python
# Identity matrix with float type
e=np.eye(3, dtype=np.float64)
print(e)
print(e.dtype)
```

# **Changing the Data Type of an Existing Array:**

            Use array.astype(new_dtype) to convert an arrays data type.

```python
a=np.array([1.2, 2.5, 3.8])
print(a)
print(a.dtype)
```

```python
# Let us convert float array into integer array
i=a.astype(np.int32)
print(i)
print(i.dtype)
```

```python
# Let us convert int array into string array
s=i.astype(np.str_)
print(s)
print(s.dtype)
```

# **Creating a NumPy DateTime Array:**

```python
import numpy as np
x=np.array(["24-06-2025"])
print(x)
print(type(x))
print(x.dtype)
```

```python
# Creating a single date
date=np.datetime64('2025-03-18')
print(date)
print(date.dtype)
```

**Specifying Time Precision:**
NumPy allows specifying different time units (Y, M, D, h, m, s, ms, us, ns).

```python
# Year precision
print(np.datetime64('2025','Y'))
```

```python
# Month precision
print(np.datetime64('2025','M')) # only year and M given - default month is 1
```

```python
# Month precision
print(np.datetime64('2025-03','M'))
```

```python
# Day precision
print(np.datetime64('2025-03-13','D'))
```

```python
# Day precision
print(np.datetime64('2025','D'))
```

```python
# Minute precision
print(np.datetime64('2024-03-12T14:30','m'))
```

```python
# Second precision
print(np.datetime64('2024-03-12T14:30:45','s'))
```

**Using timedelta:**

```python
# Add days
print(np.datetime64('2026-02-04') + np.timedelta64(10,'D'))
```

```python
# Subtract months
print(np.datetime64('2025-06') - np.timedelta64(2,'M'))
```

```python
# Add hours
print(np.datetime64('2024-03-12T12','h') + np.timedelta64(5,'h'))
```

**Finding Differences Between Dates:**

```python
d1 = np.datetime64('2025-03-13')
d2 = np.datetime64('2024-03-01')

diff = d1 - d2
print(diff)
print(diff.dtype)
```

```python
# array of Dates
dates=np.arange('2026-01-01', '2026-01-24', dtype='datetime64[D]')
print(dates)
```

```python
# Printing Current Date and Time
print(np.datetime64('now'))
```

```python
print(np.datetime64('now', 'D'))
```

```python
print(np.datetime64('now', 'h'))  # current hour
```

```python
print(np.datetime64('now', 's'))
```

# **3D arrays:**

```python
import numpy as np
```

```python
x=np.eye(2)
y=np.eye(2)*2
print(x)
print("\n")
print(y)
```

```python
z=np.array([x,y])
print(z)
print("\n")
print(z.shape)
print("\n")
print(z.ndim)
```
