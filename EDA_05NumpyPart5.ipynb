# **Loading Data From External Files:**


1.   **np.loadtxt()**: Load data from a text file.

Loads all values as strings

Skips the header row

Good to use when CSV has no missing values

2.   **np.genfromtxt()**: Load data from a text file, with missing values handled as specified.

Automatically infers numeric types means NumPy automatically figures out the data type (dtype) of the data by looking at the values, without you explicitly specifying it.

Much more flexible than loadtxt

Handles missing values

Good to use when Dataset may have missing values and Data is mostly numeric





```python
import numpy as np
```

```python
# Let us try creating array using Loadtxt()
data = np.loadtxt('/content/Iris.csv',delimiter=",",skiprows=1,dtype="str")
# all data in string format because of loadtext()
print(type(data))
print(data.shape)
print(data.dtype)
```

```python
data[:,1].mean() # try to find mean of 1st column
# error as all columns are of string data type
```

```python
# solution: 1st convert it to float then find mean
sepal_length = data[:,1].astype(float)
print(sepal_length.mean())
```

# **Matrix operations in Numpy:**


1.   matrix multiplication
2.   matrix inversion
3.   dot product and cross product
4.   eigen values and eigen vectors.



```python
# matrix multiplication:
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print(a)
print('\n')
print(b)
print('\n')
c=np.matmul(a,b)
print(c)
```

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
c=np.dot(a,b)
print(c)
```

# **Matrix Inversion**


image.png

```python
# matrix inversion
a=np.array([[1,2],[3,4]])
# square matrix, Only square matrices can have an inverse.
# square matrix = same no. of rows and columns, Linear algebra
b=np.linalg.inv(a)
print(b)
```

```python
x=np.eye(3)
y=np.linalg.inv(x)
print(y)
```

```python
# cross product.
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.cross(a,b)
print(c)
```

```python
# cross product of 2d arrays
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
c=np.cross(a,b)
print(c)
```

```python
# eigen values of a matrix
a=np.array([[1,2],[3,4]])
b,c=np.linalg.eig(a)
print(b)
print(c)
```
