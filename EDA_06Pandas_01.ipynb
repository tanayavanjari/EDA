# **Pandas Data Structures:**

Pandas provides two primary data structures that are essential for data manipulation and analysis: Series and DataFrame.



1.   **Series**

              (a) Pandas Series is a one-dimensional labeled array capable of holding any data type (integer, string, etc.)
              (b) Pandas Series is similar to a column in a table.
              (c) Pandas Series consists of an array of data and an associated array of labels, called the index.

2.   **DataFrame**

              (a) DataFrame is a two-dimensional labeled data structure with columns of potentially different data types.
              (b) It is similar to a table in a database or an Excel spreadsheet.
              (c) A DataFrame consists of an array of data and two associated arrays of labels:
                          row labels (index) and column labels.

# **Creating Pandas Series:**

Pandas series is created in many ways.



1.   from a numpy 1D array.
2.   from list of elements.
3.   from scalar value.
4.   from a python dictionary.




```python
# Import pandas library
import pandas as pd
import numpy as np
```

```python
# Creating Series using 1D array
arr=np.arange(10,51,10)
print(arr)
print(type(arr))
print("\n")
s=pd.Series(arr)
print(s)
```

```python
# Creating a Series with mutable index
print(arr)
print("\n")
s=pd.Series(arr,index=["first","second","third","fourth","fifth"])
print(s)
```

```python
# Creating  Series using list of elements.
s=pd.Series([11,22,33,44,55])
print(s)
```

```python
# Creating s Series with scalar values.
s=pd.Series(99,index=[1,2,3,4,5,6,7,8])
print(s)
```

```python
# Creating s Series with scalar values.
s=pd.Series(9,index=[1,2,3])
print(s)
```

```python
s=pd.Series(99,index=list(range(10000)))
print(s)
```

```python
s=pd.Series(list(range(10)))
print(s)
```

```python
s=pd.Series(np.inf,index=list(range(5)))
print(s)
```

```python
s=pd.Series(np.nan,index=[1,2,3,4,5])
print(s)
```

```python
# Creating a Series using python dictionary.
d={"name":"MS Dhoni","Team":"CSK","Role":"Captain","Position":7}
print(d)
print(type(d))

s=pd.Series(d)
print("\n")
print(s)
```

# **Pandas Data Frame:**



1.   It is used to represent 2D data.
2.   It is a collection of series.
3.   In data frame, data and size both are Mutable.



```python
# Creating a Data Frame:
# empty data frame
df=pd.DataFrame()
print(df)
print(type(df))
```

```python
# Converting a pandas series into a dataframe.
print(s) # s is a dictionary in above command
df=pd.DataFrame(s)
print("\n")
print(df)
```

```python
# Creating dataframe form 2D array
arr=np.arange(10,101,10).reshape(5,2)
print(arr)
df=pd.DataFrame(arr)
print("\n")
print("data frame:")
print(df)
```

```python
# Creating dataframe form 2D array
arr=np.arange(10,101,10).reshape(5,2)
print(arr)
df=pd.DataFrame(arr,columns=["col_01","col_02"])
print("\n")
print("data frame:")
print(df)
```

```python
# Creating data frame from list of dictionaries.
d=[{"name":"MS Dhoni","Team":"CSK"},{"name":"Virat","Team":"RCB"},
   {"name":"KL Rahul","Team":"LSG"}]
df=pd.DataFrame(d)
print(df)
```

```python
# Creating data frame from list of lists.
d=[["MS Dhoni","CSK"],["Virat Kohli","RCB"],["KL Rahul","LSG"]]
df=pd.DataFrame(d,columns=["Player","Team"])
print(df)
```
