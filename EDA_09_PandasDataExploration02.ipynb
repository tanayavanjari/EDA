# **items()**: It loops column by column.
# **iterrows()**: It loops row by row.

```python
# items(): It loops column by column.
# Each iteration gives:
# (column_name, column_data_as_series)
# In simple words items() says: Give me one entire column at a time.

for col_name, col_data in df.items():
    print("Column:", col_name)
    print(col_data.head())
    print("----")
```

```python
df
```

```python
# iterrows(): It loops row by row
# Each iteration gives:
# (index, row_data_as_series)
# In simple words iterrows() says: Give me one student record at a time.

for index, row in df.iterrows():
    print("Index:", index)
    print("Name:", row['name'])
    print("IT Experience:", row['it_experience'])
    print("Salary:", row['salary_expectations'])
    print("----")
```

# **12. Find Index Labels for min/max:**

               (a) label = df['col1'].idxmin()
               (b) label = df['col1'].idxmax()
               (c) df.nlargest(n,col_name)
               (d) df.nsmallest(n,col_name)

```python
df = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D', 'E'],
    'col1': [45, 12, 78, 30, 60]
})

print(df)
```

```python
# idxmin(): Returns the index label where the minimum value occurs
label = df['col1'].idxmin()
print(label)
```

```python
# idxmax(): Returns the index label where the maximum value occurs
label = df['col1'].idxmax()
print(label)
```

```python
# nlargest(): Returns top n rows with largest values in a column,
# output is a dataframe
top_n = df.nlargest(3, 'col1')
print(top_n)
```

```python
# nsmallest(): Returns top n rows with smallest values in a column
bottom_n = df.nsmallest(2, 'col1')
print(bottom_n)
```

# **Ranking Methods:**

| Method      | Description                                               | Example Ranks (for \[80, 95, 95, 70]) |
| ----------- | --------------------------------------------------------- | ------------------------------------- |
| `'average'` | (Default) Assign average of the ranks to tied values      | `[2.0, 3.5, 3.5, 1.0]`                |
| `'min'`     | Assign **lowest** possible rank to all tied values        | `[2.0, 3.0, 3.0, 1.0]`                |
| `'max'`     | Assign **highest** possible rank to all tied values       | `[2.0, 4.0, 4.0, 1.0]`                |
| `'first'`   | Assign ranks in the **order values appear**, even if tied | `[2.0, 3.0, 4.0, 1.0]`                |
| `'dense'`   | Like `min`, but ranks go up by 1 (no gaps)                | `[2.0, 3.0, 3.0, 1.0]`                |


```python
data = {
    'Eid':[100,200,300,110,120,210,220,220],
    'Ename':['Disha','Manali','Reva','Amit','Raj','Shahank','Saloni','Nisha'],
    'Job':['Mgr','Programmer','HR','Admin','Accounts','Sales','Sales','Programmer']
}
newdf = pd.DataFrame(data)
newdf
```

```python
# df.rank(): Assigns ranks to numeric values.
# For categorical values ranks are assigned alphabetically
# It ranks column-wise by default.
# If same values in a column for 2 or more rows it takes average
# So in Eid row 6 and 7 eid was 220. So rank is (6+7)/2=6.5
# same for categorical columns
newdf.rank()
```

```python
newdf['Eid'].rank(method='average')
# for rank() default argument is method='average'
# so got same result as rank()
```

```python
newdf['Eid'].rank(method='min')
# Duplicates get smallest rank.
```

```python
newdf['Eid'].rank(method='max')
# # Duplicates get highest rank.
```

```python
newdf['Eid'].rank(method='dense')
# No rank gaps
```

```python
newdf['Eid'].rank(method='first')
# Ranks assigned in order of appearance.
```

```python
newdf.sort_values(by='Eid')
# Sort rows based on column values.
```

```python
newdf.sort_values(by=['Job','Eid'])
# Sort by multiple columns
```
