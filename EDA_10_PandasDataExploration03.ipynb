## Filtering Rows from a DataFrame
### Way 1
We can select the rows by using slicing operation.
### Syntax: df[ starting_row_index : ending_row_index : step ]

### Way 2
Similar to numpy Pandas can accept boolean indexes.
To select rows based on a conditional expression, use a condition inside the selection brackets [].
### Syntax: df[ CONDITION ]
- df[ df["maximum temperature"] > 95 ]
- df[df['date'].isin(['10-5-2016', '10-4-2016', '10-6-2016'])]
- df[ (df['date']=='10-5-2016') |
    (df['date']=='10-4-2016') |
    (df['date']=='10-6-2016')
  ]

##  How to resolve this? 
In case, you want a subset of both rows and columns in one go, just using selection brackets [] is not sufficient anymore.
Here loc/iloc operators are required in front of the selection brackets []. When using loc/iloc, the part before the comma is the rows you want, and the part after the comma is the columns you want to select.

## Syntax:
- df.loc[row_label, col_label]
- df.iloc[row_index, col_index]

```python
# Row number 4
df.loc[4]     # index base access
```

```python
df.loc[1:4]
```

```python
df.loc[1:4,'NAME'] # only one column, column label can be written
```

```python
df.loc[1:4,['NAME','BATCH NUMBER']]
```

```python
df.loc[1:4,'NAME':'COURSE REGISTERED']
```

```python
df.loc[1:4,2] # column index not allowed, only column label can be written
```

```python
df.iloc[1:4,2]
```

```python
df.iloc[1:4,2:5]
```

```python
df.columns
```

####  Accessing rows based on a condition
- df.loc[CONDITION , col_lables ]

####  Accessing rows based on multiple condition
df.loc[ (COND_1) & (COND_2) | (COND_3) , col_lables ]

### condition based accessing without loc

```python
df[df['FIELD OF STUDY']=='Computer Science']['NAME']
```

# Multiple Conditions without using loc

```python
# all columns are visible for thos record
df[(df['BATCH NUMBER']==396) & (df['HOMETOWN']=='Hyderabad')]
```

```python
# for the same record we want only Name column to be displayed
df[(df['BATCH NUMBER']==396) & (df['HOMETOWN']=='Hyderabad')]['NAME']
```

```python
# Another way: for the same record we want only Name column to be displayed
df.loc[(df['BATCH NUMBER']==396)&(df['HOMETOWN']=='Hyderabad'),'NAME']
```

```python
# for the same record we want Name and Highest Qualifications columns to be displayed
df.loc[(df['BATCH NUMBER']==396)&(df['HOMETOWN']=='Hyderabad'),['NAME','HIGHEST QUALIFICATION']]
```
