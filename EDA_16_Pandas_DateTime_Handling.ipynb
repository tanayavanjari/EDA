## date_range()
- ##### pandas.date_range() is one of the general functions in Pandas which is used to return a fixed frequency DatetimeIndex.

```python
# Day wise
pd.date_range("27/03/2025", periods = 7) # output format: yyyy/mm/dd
```

```python
# Month wise
print(pd.date_range("03/27/2025", periods = 5, freq = "ME"))
```

```python
#Month wise MS- Month start
print(pd.date_range("03/10/2025", periods = 5, freq = "MS"))
```

```python
# Year wise
print(pd.date_range("03/27/2025", periods = 6, freq="YE"))
```

```python
# Year wise
print(pd.date_range("01/08/2025", periods = 6, freq="YS"))
```

```python
#Create  a datetime datatype with specified intervals

start = pd.to_datetime("2025-1-1")    # (Y, M, D)
print("Start: ",start) #OR
# print(f"Start : {start}")

end = pd.to_datetime("2025-2-20")
print("End : ",end)

pd.date_range(start, end)
```
