# **Data Vizualization:**

Data visualization is the process of representing data graphically or visually, making complex information easier to understand and interpret. It transforms raw data into visual formats like charts, graphs, maps, and infographics, allowing patterns, trends, and insights to emerge at a glance.

**Importance of Data Visualization**

**(a) Improves Understanding:**
Visual representations simplify complex data sets, enabling faster comprehension.

For **example**, a line chart can effectively showcase sales growth over time compared to reading raw numbers.

**(b) Identifies Trends and Patterns:**

Helps reveal relationships, anomalies, or trends in the data that might not be obvious from tables or text.

**Example:** Heatmaps highlight areas with higher activity or intensity.

**(c) Supports Decision Making:**
Decision-makers rely on visuals for quick, informed choices based on data insights.

For **instance**, dashboards presenting KPIs (Key Performance Indicators) enable managers to track performance efficiently.

**(d) Engages and Communicates Effectively:**

Visuals are more engaging and memorable compared to raw data, making it easier to convey ideas and findings.

**Example:** Pie charts effectively break down proportions in survey result

# **Key Components of Plot:**

**(a) Title:**

Describes the purpose or context of the plot.

**Example:** "Monthly Sales Trend in 2023".

**(b) Axes:**

*X-Axis (Horizontal):*
Represents the independent variable (e.g., time, categories).

*Y-Axis (Vertical):*

Represents the dependent variable (e.g., sales, frequency).

**(c) Scale and Tick Marks:**
Indicates the range and intervals along the axes.

**Example:** Time intervals (e.g., months on the x-axis) or numeric increments (e.g., revenue in thousands on the y-axis).

**(d) Labels:**
Axes should have clear labels describing what the variables represent, often with units.

**Example:** "Revenue ($)" on the y-axis, "Month" on the x-axis.

**(e) Data Representation:**

The graphical element that represents the data:

**Points:** For scatter plots.
**Lines:** For trends in line charts.
**Bars:** For comparisons in bar charts.
**Legend (Optional):**

Explains the meaning of different colors, lines, or symbols if there are multiple data series.

**Example:** A legend indicating "Product A" in blue and "Product B" in orange.

**(f) Gridlines (Optional):**

Horizontal and/or vertical lines to help align data points with their corresponding axes.

**(g) Annotations (Optional):**

Highlights or explanations for specific data points or trends within the plot.

**Example:** "Sales Spike in December" with an arrow pointing to the peak.

**(h) Color/Style:**
Visual style enhances readability and emphasizes important elements:

Colors for categories or intensities (e.g., heatmaps).
Line thickness or patterns for differentiation.

# **How to Read a Plot:**

1. Start with the title to understand the context.
2. Look at the axes labels to identify the variables being compared.
3. Observe the data points or lines to identify trends, patterns, or anomalies.
4. Refer to the legend for interpretation, if multiple series or categories exist.

```python
# Line Plot With Custom Tiks
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create line plot
plt.plot(x, y, label='y = 2x')

# Add custom ticks for x and y axes
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['One', 'Two', 'Three', 'Four', 'Five'])
plt.yticks(ticks=[2, 4, 6, 8, 10], labels=['Two', 'Four', 'Six', 'Eight', 'Ten'])

# Add labels, title, legend, and grid
plt.title("Line Plot with Custom Ticks")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid()
plt.show()
```

```python
# Adding Multiple Lines
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y, label='y = 2x', color='blue')
plt.plot(x, y2, label='y2 = 2x - 1', color='red', linestyle=':')
plt.title("Line Plot with Multiple Lines")
plt.xlabel("X-axis Data")
plt.ylabel("Y-axis Data")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x, y, color='blue')
plt.plot(x, y2, color='red', linestyle=':')
plt.title("Line Plot with Multiple Lines")
plt.xlabel("X-axis Data")
plt.ylabel("Y-axis Data")
plt.legend(["x-y","x-y2"])
plt.grid()
plt.show()
```

```python
# Creating Multiple Plots
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 3, 4, 5]
y3 = [5, 4, 3, 2, 1]
y4 = [10, 8, 6, 4, 2]

# Subplot: 2x2 grid subplot(r,c,position)
plt.subplot(2, 2, 1)  # First plot (top-left)
plt.plot(x, y1, color='red', label='y1 = 2x')
plt.title("Plot 1")
plt.legend()

plt.subplot(2, 2, 2)  # Second plot (top-right)
plt.plot(x, y2, color='blue', label='y2 = x')
plt.title("Plot 2")
plt.legend()

plt.subplot(2, 2, 3)  # Third plot (bottom-left)
plt.plot(x, y3, color='green', label='y3 = reverse(x)')
plt.title("Plot 3")
plt.legend()

plt.subplot(2, 2, 4)  # Fourth plot (bottom-right)
plt.plot(x, y4, color='purple', label='y4 = 10-2x')
plt.title("Plot 4")
plt.legend()

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
# Subplot: 2x2 grid
plt.subplot(2, 2, 1)  # First plot (top-left)
plt.scatter(x, y1, color='red', label='y1 = 2x')
plt.title("Plot 1")
plt.legend()

plt.subplot(2, 2, 2)  # Second plot (top-right)
plt.plot(x, y2, color='blue', label='y2 = x')
plt.title("Plot 2")
plt.legend()

plt.subplot(2, 2, 3)  # Third plot (bottom-left)
plt.scatter(x, y3, color='green', label='y3 = reverse(x)')
plt.title("Plot 3")
plt.legend()

plt.subplot(2, 2, 4)  # Fourth plot (bottom-right)
plt.plot(x, y4, color='purple', label='y4 = 10-2x')
plt.title("Plot 4")
plt.legend()

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]
y2 = [30, 25, 20, 15, 10]

# Create 2 plots in a 2x1 grid
fig, axes = plt.subplots(2, 1, figsize=(6, 8))  # 2 rows, 1 column
# fig object represents entire image (like a full page)
# We can use fig object when we want to
# Change overall size, Add a main title
# Save the whole figure, Adjust spacing

# axes object represents the individual plotting areas (The boxes drawn on the page)

# First plot
axes[0].plot(x, y1, color='orange', label='Increasing Trend')
axes[0].set_title("Top Plot")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].legend()
axes[0].grid()

# Second plot
axes[1].plot(x, y2, color='green', label='Decreasing Trend')
axes[1].set_title("Bottom Plot")
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Y-axis")
axes[1].legend()
axes[1].grid()

# Adjust layout and display
plt.tight_layout()
fig.savefig("plot.png")
plt.show()
```

```python
# plt.figure(1) and plt.figure(2) creates 2 completely separate figures, not subplots inside one figure.
# if not mentioned it will overlap 2 figures in one cell. try commenting plt.figure(1) and plt.figure(2)

# First figure
plt.figure(1) # It creates or activates a new Figure window.
plt.plot(x, y1, color='blue', label='Figure 1')
plt.title("Figure 1")
plt.legend()

# Second figure
plt.figure(2)
plt.plot(x, y2, color='red', label='Figure 2')
plt.title("Figure 2")
plt.legend()

plt.show()
```
